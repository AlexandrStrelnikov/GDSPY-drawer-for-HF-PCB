```python
imp.reload(gdspy)
from element import *
from chip_element import *
from sketch import *

S = 78
W = 47
a = 16.5e3

sketch = Sketch('20-pin sample holder')
baseGeometry = gdspy.Rectangle(((a+0.6)/2, (a+0.6)/2), (-(a+0.6)/2, -(a+0.6)/2))
baseGeometry = gdspy.boolean(baseGeometry, gdspy.Rectangle(((a-0.6)/2, (a-0.6)/2), (-(a-0.6)/2, -(a-0.6)/2)), 'not')
sketch.boolean(baseGeometry, 'or', 0)

portePerEdge = 5
portDisp = 1280*2

ports = []
for j in range(0, 4):
    portsEdge = []
    for i in range(0, portePerEdge):
        port = Port((-a/2, -(portePerEdge-1)/2*portDisp), 0)
        port.translate((0, portDisp*i))
        port.rotate(j*np.pi/2, (0,0))
        
        chipPort = ChipCPWPort(base_length = 900, translation_length = 600, S = S, W = W)
        
        # переопределяем в случае, когда нужны другие линии
        if j == 1 or (j == 0 and i in [1, 2]) or (j == 2 and i == 3):
            chipPort = ChipCPWPort(base_length = 900, translation_length = 600, S = 25.2, W = 10.8)
        if (j == 0 and i == 3) or (j == 2 and i == 1):
            chipPort = ChipCPWPort(base_length = 900, translation_length = 600, S = 25.2, W = 10.8)
        
            
            
        chipPort.mergeWithPort(port, 0)
        sketch.addGeometry(chipPort)
        portsEdge += [chipPort.ports[1]]
    ports += [portsEdge]
    mark = SawMark((-a/2,-a/2))
    mark.rotate(np.pi/2*j, (0, 0))
    sketch.addGeometry(mark)
    

    
cpw = CPW_Chip(ports[0][4],ports[2][0], S, W, r = 1000)
sketch.addGeometry(cpw)

# Рисуем резонаторы
base_cavity = NotchQuarterCavity(tail = 100)
base_cavity.move((-604.8*2, 1.28e3*2))
base_cavity.rotate(np.pi)
cavities = [base_cavity]
sketch.addGeometry(base_cavity)
for i in range(1,4):
    cave = NotchQuarterCavity(tail = 200 + 150*i)
    cave.mergeWithPort(cavities[-1].ports[0], 1)
    cavities += [cave]
    sketch.addGeometry(cave)

# добавляем обычный резонатор
cave = NotchQuarterCavity(tail = 0)
cave.mergeWithPort(cavities[-1].ports[0], 1)
cavities += [cave]
sketch.addGeometry(cave)
ac_end = ACEnd(S = 21.2, W = 8.8, length = 30)
ac_end.mergeWithPort(cavities[-1].ports[2], 0)
sketch.addGeometry(ac_end)

# соединяем резонаторы с линиями
cpw = CPW_Chip(cavities[-1].ports[0],ports[2][1], 25.2, 10.8, r = 1000)
sketch.addGeometry(cpw)

cpw = CPW_Chip(ports[0][3],cavities[0].ports[1], 25.2, 10.8, r = 1000)
sketch.addGeometry(cpw)

# рисуем кресты
crosses = []
for i in range(0, 4):
    cross = Cross(a=400+20*i)
    cross.mergeWithPort(cavities[i].ports[2], 0)
    crosses += [cross]
    sketch.addGeometry(cross)

# рисуем джозики
squids = []
for i in range(0,4):
    squid = SQUID_3point()
    squid.mergeWithPort(crosses[i].ports[1], 0)
    squids += [squid]
    sketch.addGeometry(squid)
    
ac_ends = []
dc_ends = []
for i in range(0,4):
    ac_end = ACEnd()
    dc_end = BiasEnd()
    ac_end.mergeWithPort(crosses[i].ports[3], 0)
    dc_end.mergeWithPort(crosses[i].ports[2], 0)
    ac_ends += [ac_end]
    dc_ends += [dc_end]
    sketch.addGeometry(ac_end)
    sketch.addGeometry(dc_end)

    
# соединяем линии драйва
# первый кубит 
# здесь ломается определитель рабиуса!
cpw = CPW_Chip(ports[0][2], ac_ends[0].ports[1], 25.2, 10.8, r = 500)
sketch.addGeometry(cpw)
cpw = CPW_Chip(ports[0][1], dc_ends[0].ports[1], 25.2, 10.8, r = 3000)
sketch.addGeometry(cpw)

# второй кубит 
cpw = CPW_Chip(ports[1][4], ac_ends[1].ports[1], 25.2, 10.8, r = 3500)
sketch.addGeometry(cpw)
cpw = CPW_Chip(ports[1][3], dc_ends[1].ports[1], 25.2, 10.8, r = 2000)
sketch.addGeometry(cpw)

# третий кубит 
cpw = CPW_Chip(ports[1][2], ac_ends[2].ports[1], 25.2, 10.8, r = 4000)
sketch.addGeometry(cpw)
cpw = CPW_Chip(ports[1][1], dc_ends[2].ports[1], 25.2, 10.8, r = 4000)
sketch.addGeometry(cpw)

# четвертый кубит 
cpw = CPW_Chip(ports[2][3], dc_ends[3].ports[1], 25.2, 10.8, r = 3000)
sketch.addGeometry(cpw)
cpw = CPW_Chip(ports[1][0], ac_ends[3].ports[1], 25.2, 10.8, r = 4000)
sketch.addGeometry(cpw)

#наносим метки совмещения
mark = LitographyMark(position = (-14e3/2, -14e3/2))
sketch.addGeometry(mark)

mark = LitographyMark(position = (14e3/2, -14e3/2))
sketch.addGeometry(mark)

mark = LitographyMark(position = (-14e3/2, 14e3/2))
sketch.addGeometry(mark)

mark = LitographyMark(position = (14e3/2, 14e3/2))
sketch.addGeometry(mark)

mark = LitographyMark(position = (0, 4e3))
sketch.addGeometry(mark)


    
# mark = LitographyMark()
# sketch.addGeometry(mark)

# cpw = CPW_Chip(ports[0][2],dc_ends[1], S, W, r = 1000)
# sketch.addGeometry(cpw)

# cpw = CPW_Chip(ports[0][2],ports[2][2], S, W, r = 1000)
# sketch.addGeometry(cpw)

# loop_len = 6e3
# cpw = CPW_Chip(ports[0][1], ports[0][0], S, W, r = 500,
#                path = [ports[0][1].position, np.array(ports[0][1].position) + np.array([loop_len,0]),
#                        np.array(ports[0][0].position) + np.array([loop_len,0]), ports[0][0].position])
# sketch.addGeometry(cpw)

# cpw = CPW_Chip(ports[2][4], ports[2][3], S, W, r = 500,
#                path = [ports[2][4].position, np.array(ports[2][4].position) - np.array([loop_len,0]),
#                        np.array(ports[2][3].position) - np.array([loop_len,0]), ports[2][3].position])
# sketch.addGeometry(cpw)

sketch.draw()
sketch.saveGDS('20pin flip chip base with qubits')
```
